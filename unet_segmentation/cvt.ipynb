```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
for i in range(1,21):
# 读取图片
    img_jpg = cv2.imread(f"./DRIVE/img/{i}.jpg")
    img_png = cv2.imread(f"./DRIVE/img/{i}.png")

# 变换

    # 上下翻转，左右不动
    x_jpg = cv2.flip(img_jpg, 0)
    x_png = cv2.flip(img_png, 0)
    # 左右翻转，上下不动
    y_jpg = cv2.flip(img_jpg, 1)
    y_png = cv2.flip(img_png, 1)
    # 上下左右都翻转
    xy_jpg = cv2.flip(img_jpg, -1)
    xy_png = cv2.flip(img_png, -1)  

    cv2.imwrite(f"./img_cvt/{4*i-3}.jpg", img_jpg)
    cv2.imwrite(f"./img_cvt/{4*i-2}.jpg", x_jpg)
    cv2.imwrite(f"./img_cvt/{4*i-1}.jpg", y_jpg)
    cv2.imwrite(f"./img_cvt/{4*i}.jpg", xy_jpg)
    cv2.imwrite(f"./img_cvt/{4*i-3}.png", img_png)
    cv2.imwrite(f"./img_cvt/{4*i-2}.png", x_png)
    cv2.imwrite(f"./img_cvt/{4*i-1}.png", y_png)
    cv2.imwrite(f"./img_cvt/{4*i}.png", xy_png)
    
    
```

```python
img = cv2.imread("./DRIVE/img/1.jpg")
    # 上下翻转，左右不动
x = cv2.flip(img, 0)
    # 左右翻转，上下不动
y = cv2.flip(img, 1)
    # 上下左右都翻转
xy = cv2.flip(img, -1)  
```

```python
plt.subplot(2, 2, 1)
plt.title("img")
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))

plt.subplot(2, 2, 2)
plt.title("x")
plt.imshow(cv2.cvtColor(x, cv2.COLOR_BGR2RGB))

plt.subplot(2, 2, 3)
plt.title("y")
plt.imshow(cv2.cvtColor(y, cv2.COLOR_BGR2RGB))

plt.subplot(2, 2, 4)
plt.title("xy")
plt.imshow(cv2.cvtColor(xy, cv2.COLOR_BGR2RGB))
```
